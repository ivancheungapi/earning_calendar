```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('events.xlsx')
df
```

```python
event_df = df[['Name', 'Ticker', 'Date', 'Event Type' , 'Period']]
event_df
```

```python
def clean_period(period):

    if period is None or pd.isna(period):
        return period
    else:
        period = str(period)
        if '(' in period:
            split_period = period.split('(')[0].strip()
            return split_period
        else:
            return period
event_df['Period'] = event_df['Period'].apply(clean_period)
event_df
```

```python
#drop duplicates
event_df = event_df.drop_duplicates()
event_df
```

```python
def full_event_name(abbreviation):
    if abbreviation == 'SR':
        return 'Sales Release'
    elif abbreviation == 'ER':
        return 'Earnings Release'
    elif abbreviation == 'EC':
        return 'Earning Call'
    else:
        return abbreviation
event_df['Event Type'] = event_df['Event Type'].apply(full_event_name)
```

```python
event_df['Time'] = event_df['Date'].dt.strftime('%H:%M')
```

```python
import calendar

# Ensure Date is datetime
event_df["Date"] = pd.to_datetime(event_df["Date"]).dt.date

def build_month_calendar_html(events: pd.DataFrame, year: int, month: int, title: str = "Earnings Calendar") -> str:
    # Prepare events lookup: {date: [rows...]}
    month_events = events[
        (pd.to_datetime(events["Date"]).dt.year == year) &
        (pd.to_datetime(events["Date"]).dt.month == month)
    ].copy()

    by_day = {}
    for _, r in month_events.iterrows():
        d = r["Date"]
        by_day.setdefault(d, []).append(r)

    # Calendar matrix (weeks x days) with Monday as first day
    cal = calendar.Calendar(firstweekday=0)  # 0=Monday
    weeks = cal.monthdatescalendar(year, month)

    month_name = calendar.month_name[month]

    # Minimal CSS + HTML
    html = f"""<!doctype html>
<html lang="en">
<head>
  <meta charset="utf-8" />
  <meta name="viewport" content="width=device-width,initial-scale=1" />
  <title>{title} â€” {month_name} {year}</title>
  <style>
    :root {{
      --bg: #0b0f14;
      --card: #121823;
      --muted: #9aa4b2;
      --text: #e8edf3;
      --border: #223043;
      --chip: #1a2433;
      --chip-border: #2a3a52;
      --accent: #6aa6ff;
    }}
    body {{
      margin: 0;
      font-family: system-ui, -apple-system, Segoe UI, Roboto, Arial, sans-serif;
      background: var(--bg);
      color: var(--text);
    }}
    .wrap {{
      max-width: 2000px;
      margin: 0 auto;
      padding: 36px 24px 56px;
    }}
    header {{
      display: flex;
      gap: 16px;
      align-items: baseline;
      justify-content: space-between;
      margin-bottom: 22px;
    }}
    h1 {{
      font-size: 26px;
      margin: 0;
      font-weight: 750;
      letter-spacing: 0.2px;
    }}
    .sub {{
      color: var(--muted);
      font-size: 14px;
    }}
    .grid {{
      display: grid;
      grid-template-columns: repeat(7, 1fr);
      gap: 14px;
    }}
    .dow {{
      color: var(--muted);
      font-size: 13px;
      text-transform: uppercase;
      letter-spacing: 0.10em;
      padding: 0 8px;
    }}
    .cell {{
      background: var(--card);
      border: 1px solid var(--border);
      border-radius: 14px;

      min-height: 165px;    
      max-height: 320px;      

      padding: 12px;
      display: flex;
      flex-direction: column;
      gap: 10px;

      overflow: hidden;       
    }}
    .cell.other-month {{
      opacity: 0.45;
    }}
    .daynum {{
      display: flex;
      justify-content: space-between;
      align-items: center;
      font-size: 13px;
      color: var(--muted);
    }}
    .daynum .n {{
      font-weight: 800;
      color: var(--text);
      font-size: 14px;
    }}
    .chips {{
      display: flex;
      flex-direction: column;
      gap: 8px;

      overflow-y: auto;     
      padding-right: 6px;    
    }}
    .chip {{
      background: var(--chip);
      border: 1px solid var(--chip-border);
      border-radius: 10px;
      padding: 8px 10px;      /* was 10px 10px */
      line-height: 1.2;
    }}
    .chip .top {{
      display: flex;
      justify-content: space-between;
      gap: 10px;
      align-items: baseline;
    }}
    .ticker {{
      font-weight: 800;
      color: var(--accent);
      font-size: 13px;
      white-space: nowrap;
    }}
    .name {{
      font-size: 13px;
      color: var(--text);
      overflow: hidden;
      text-overflow: ellipsis;
      white-space: nowrap;
      flex: 1;
    }}
    .meta {{
      margin-top: 4px;
      color: var(--muted);
      font-size: 11px;
      display: flex;
      gap: 8px;
      flex-wrap: nowrap;
      white-space: nowrap;
    }}
    .pill {{
      border: 1px solid var(--border);
      border-radius: 999px;
      padding: 3px 10px;
    }}
  </style>
</head>
<body>
  <div class="wrap">
    <header>
      <div>
        <h1>{title}</h1>
        <div class="sub">{month_name} {year}</div>
      </div>
      <div class="sub">Source: Bloomberg</div>
    </header>

    <div class="grid" aria-label="calendar">
      {"".join([f'<div class="dow">{d}</div>' for d in ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]])}
    </div>

    <div class="grid" style="margin-top:10px;">
"""

    for week in weeks:
        for d in week:
            is_other = (d.month != month)
            day_events = by_day.get(d, [])
            cell_class = "cell other-month" if is_other else "cell"

            chips_html = ""
            for r in day_events:
                # Safe string conversions
                name = str(r.get("Name", "") or "")
                ticker = str(r.get("Ticker", "") or "")
                etype = str(r.get("Event Type", "") or "")
                period = str(r.get("Period", "") or "")
                time_str = str(r.get("Time", "") or "")

                chips_html += f"""
        <div class="chip">
          <div class="top">
            <div class="ticker">{ticker}</div>
            <div class="name" title="{name}">{name}</div>
          </div>
          <div class="meta">
            <span class="pill">{period} {etype}</span>
          <span class="pill">{time_str}</span>

          </div>
        </div>
"""

            html += f"""
      <div class="{cell_class}">
        <div class="daynum">
          <span class="n">{d.day}</span>
          <span>{d.strftime("%Y-%m-%d")}</span>
        </div>
        <div class="chips">
          {chips_html if chips_html else '<div class="sub"> </div>'}
        </div>
      </div>
"""

    html += """
    </div>
  </div>
</body>
</html>
"""
    return html
```

```python
# Pick a month to render (e.g., the min date month in your data)
min_d = min(event_df["Date"])
year, month = min_d.year, min_d.month

html = build_month_calendar_html(event_df, year=year, month=month, title="Earnings Calendar")

with open("earnings_calendar.html", "w", encoding="utf-8") as f:
    f.write(html)

print("Wrote: earnings_calendar.html")
```
