```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('events1.xlsx')
df
```

```python
event_df = df[['Name', 'Ticker', 'Date', 'Event Type' , 'Period']]
event_df
```

```python
def clean_period(period):

    if period is None or pd.isna(period):
        return period
    else:
        period = str(period)
        if '(' in period:
            split_period = period.split('(')[0].strip()
            return split_period
        else:
            return period
event_df['Period'] = event_df['Period'].apply(clean_period)
event_df
```

```python
#drop duplicates
event_df = event_df.drop_duplicates()
event_df
```

```python
def full_event_name(abbreviation):
    if abbreviation == 'SR':
        return 'Sales Release'
    elif abbreviation == 'ER':
        return 'Earnings Release'
    elif abbreviation == 'EC':
        return 'Earning Call'
    else:
        return abbreviation
event_df['Event Type'] = event_df['Event Type'].apply(full_event_name)
```

```python
event_df['Time'] = event_df['Date'].dt.strftime('%H:%M')
```

```python
import json

def build_single_month_js_calendar_html(events: pd.DataFrame, title: str = "Earnings Calendar") -> str:
    # Ensure Date is date-like string for JS keys (YYYY-MM-DD)
    # (Does not change your df; we build a payload copy.)
    payload_df = events.copy()
    payload_df["DateKey"] = pd.to_datetime(payload_df["Date"]).dt.strftime("%Y-%m-%d")

    # Minimal payload for the browser
    payload = payload_df[["Name", "Ticker", "DateKey", "Event Type", "Period", "Time"]].to_dict("records")
    payload_json = json.dumps(payload, ensure_ascii=False)

    html = f"""<!doctype html>
<html lang="en">
<head>
  <meta charset="utf-8" />
  <meta name="viewport" content="width=device-width,initial-scale=1" />
  <title>{title}</title>
  <style>
    :root {{
      --bg: #0b0f14;
      --card: #121823;
      --muted: #9aa4b2;
      --text: #e8edf3;
      --border: #223043;
      --chip: #1a2433;
      --chip-border: #2a3a52;
      --accent: #6aa6ff;
    }}
    body {{
      margin: 0;
      font-family: system-ui, -apple-system, Segoe UI, Roboto, Arial, sans-serif;
      background: var(--bg);
      color: var(--text);
    }}
    .wrap {{
      max-width: 2000px;
      margin: 0 auto;
      padding: 36px 24px 56px;
    }}
    header {{
      display: flex;
      gap: 16px;
      align-items: baseline;
      justify-content: space-between;
      margin-bottom: 16px;
    }}
    h1 {{
      font-size: 26px;
      margin: 0;
      font-weight: 750;
      letter-spacing: 0.2px;
    }}
    .sub {{
      color: var(--muted);
      font-size: 14px;
    }}

    /* Month controls */
    .controls {{
      display: flex;
      align-items: center;
      justify-content: space-between;
      margin: 10px 0 16px;
      gap: 12px;
    }}
    .month-label {{
      font-size: 18px;
      font-weight: 750;
      color: var(--text);
    }}
    .btns {{
      display: flex;
      gap: 10px;
    }}
    .btn {{
      background: rgba(255,255,255,0.04);
      border: 1px solid rgba(255,255,255,0.10);
      color: var(--text);
      padding: 8px 12px;
      border-radius: 10px;
      cursor: pointer;
      font-size: 13px;
    }}
    .btn:hover {{
      background: rgba(255,255,255,0.08);
    }}

    .grid {{
      display: grid;
      grid-template-columns: repeat(7, 1fr);
      gap: 14px;
    }}
    .dow {{
      color: var(--muted);
      font-size: 13px;
      text-transform: uppercase;
      letter-spacing: 0.10em;
      padding: 0 8px;
    }}
    .cell {{
      background: var(--card);
      border: 1px solid var(--border);
      border-radius: 14px;

      min-height: 165px;
      max-height: 320px;

      padding: 12px;
      display: flex;
      flex-direction: column;
      gap: 10px;

      overflow: hidden;
    }}
    .cell.other-month {{
      opacity: 0.45;
    }}
    .daynum {{
      display: flex;
      justify-content: space-between;
      align-items: center;
      font-size: 13px;
      color: var(--muted);
    }}
    .daynum .n {{
      font-weight: 800;
      color: var(--text);
      font-size: 14px;
    }}
    .chips {{
      display: flex;
      flex-direction: column;
      gap: 8px;

      overflow-y: auto;
      padding-right: 6px;
    }}
    .chip {{
      background: var(--chip);
      border: 1px solid var(--chip-border);
      border-radius: 10px;
      padding: 8px 10px;
      line-height: 1.2;
    }}
    .chip .top {{
      display: flex;
      justify-content: space-between;
      gap: 10px;
      align-items: baseline;
    }}
    .ticker {{
      font-weight: 800;
      color: var(--accent);
      font-size: 13px;
      white-space: nowrap;
    }}
    .name {{
      font-size: 13px;
      color: var(--text);
      overflow: hidden;
      text-overflow: ellipsis;
      white-space: nowrap;
      flex: 1;
    }}
    .meta {{
      margin-top: 4px;
      color: var(--muted);
      font-size: 11px;
      display: flex;
      gap: 8px;
      flex-wrap: nowrap;
      white-space: nowrap;
    }}
    .pill {{
      border: 1px solid var(--border);
      border-radius: 999px;
      padding: 3px 10px;
    }}

    /* Themed scrollbar */
    .chips::-webkit-scrollbar {{ width: 6px; }}
    .chips::-webkit-scrollbar-track {{ background: transparent; }}
    .chips::-webkit-scrollbar-thumb {{
      background-color: rgba(255, 255, 255, 0.12);
      border-radius: 8px;
      border: 1px solid rgba(255, 255, 255, 0.08);
    }}
    .chips::-webkit-scrollbar-thumb:hover {{
      background-color: rgba(255, 255, 255, 0.30);
    }}
    .chips {{ scrollbar-width: thin; scrollbar-color: rgba(255,255,255,0.25) transparent; }}
  </style>
</head>
<body>
  <div class="wrap">
    <header>
      <div>
        <h1>{title}</h1>
        <div class="sub">Source: Bloomberg</div>
      </div>
      <div class="sub" id="summary"></div>
    </header>

    <div class="controls">
      <div class="month-label" id="monthLabel"></div>
      <div class="btns">
        <button class="btn" id="prevBtn">Prev</button>
        <button class="btn" id="nextBtn">Next</button>
      </div>
    </div>

    <div class="grid" id="dowRow" aria-label="calendar-dow"></div>
    <div class="grid" id="calGrid" style="margin-top:10px;"></div>
  </div>

<script>
  // Embedded events
  const EVENTS = {payload_json};

  // Build map by day (YYYY-MM-DD -> array of events)
  const byDay = new Map();
  const monthSet = new Set();

  for (const e of EVENTS) {{
    const dateKey = e.DateKey;
    if (!byDay.has(dateKey)) byDay.set(dateKey, []);
    byDay.get(dateKey).push(e);

    // collect unique months
    const ym = dateKey.slice(0, 7); // YYYY-MM
    monthSet.add(ym);
  }}

  // Sort months ascending
  const months = Array.from(monthSet).sort(); // "YYYY-MM"
  let idx = 0;

  // DOM nodes
  const monthLabel = document.getElementById("monthLabel");
  const calGrid = document.getElementById("calGrid");
  const dowRow = document.getElementById("dowRow");
  const summary = document.getElementById("summary");

  // Render DOW header once (Mon..Sun)
  const DOW = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"];
  dowRow.innerHTML = DOW.map(d => `<div class="dow">${{d}}</div>`).join("");

  // Helper: format month label
  function monthName(year, month1to12) {{
    const dt = new Date(year, month1to12 - 1, 1);
    return dt.toLocaleString(undefined, {{ month: "long", year: "numeric" }});
  }}

  // Helper: to YYYY-MM-DD
  function ymd(d) {{
    const y = d.getFullYear();
    const m = String(d.getMonth() + 1).padStart(2, "0");
    const day = String(d.getDate()).padStart(2, "0");
    return `${{y}}-${{m}}-${{day}}`;
  }}

  // Calendar with Monday as first day:
  // JS getDay(): Sun=0, Mon=1... We want Mon=0...Sun=6
  function mondayIndex(jsDay) {{
    return (jsDay + 6) % 7;
  }}

  function renderMonth(ym) {{
    const [yStr, mStr] = ym.split("-");
    const year = Number(yStr);
    const month = Number(mStr); // 1..12

    monthLabel.textContent = monthName(year, month);

    // Determine grid start: Monday on/before first-of-month
    const first = new Date(year, month - 1, 1);
    const start = new Date(first);
    start.setDate(first.getDate() - mondayIndex(first.getDay()));

    // Determine grid end: Sunday on/after last-of-month
    const last = new Date(year, month, 0);
    const end = new Date(last);
    end.setDate(last.getDate() + (6 - mondayIndex(last.getDay())));

    // Build days array
    const days = [];
    for (let d = new Date(start); d <= end; d.setDate(d.getDate() + 1)) {{
      days.push(new Date(d));
    }}

    // Count events this month
    let monthEventCount = 0;
    for (const [k, arr] of byDay.entries()) {{
      if (k.startsWith(ym)) monthEventCount += arr.length;
    }}
    summary.textContent = `${{monthEventCount}} events`;

    // Render cells
    calGrid.innerHTML = days.map(d => {{
      const key = ymd(d);
      const inMonth = (d.getMonth() === (month - 1));
      const cellClass = inMonth ? "cell" : "cell other-month";

      const dayEvents = byDay.get(key) || [];
      const chips = dayEvents.map(ev => {{
        const ticker = String(ev.Ticker || "").split(" ")[0];
        const name = String(ev.Name || "");
        const etype = String(ev["Event Type"] || "");
        const period = String(ev.Period || "");
        const time = String(ev.Time || "");

        return `
          <div class="chip">
            <div class="top">
              <div class="ticker">${{ticker}}</div>
              <div class="name" title="${{name.replaceAll('"','&quot;')}}">${{name}}</div>
            </div>
            <div class="meta">
              <span class="pill">${{period}} ${{etype}}</span>
              <span class="pill">${{time}}</span>
            </div>
          </div>
        `;
      }}).join("");

      return `
        <div class="${{cellClass}}">
          <div class="daynum">
            <span class="n">${{d.getDate()}}</span>
            <span>${{key}}</span>
          </div>
          <div class="chips">
            ${{chips || '<div class="sub"> </div>'}}
          </div>
        </div>
      `;
    }}).join("");
  }}

  function go(delta) {{
    idx = Math.max(0, Math.min(months.length - 1, idx + delta));
    renderMonth(months[idx]);
  }}

  document.getElementById("prevBtn").addEventListener("click", () => go(-1));
  document.getElementById("nextBtn").addEventListener("click", () => go(1));

  // Initial render: earliest month in your data
  renderMonth(months[idx]);
</script>
</body>
</html>
"""
    return html
```

```python
html = build_single_month_js_calendar_html(event_df, title="Earnings Calendar")

with open("earnings_calendar_one_month_nav.html", "w", encoding="utf-8") as f:
    f.write(html)

print("Wrote: earnings_calendar_one_month_nav.html")
```
